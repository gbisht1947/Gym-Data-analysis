```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/Dell/Desktop/python/gym.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df["Total Revenue (USD)"]=df["Total Revenue (USD)"].str[1:]
df["Annual Revenue Per Gym"]=df["Annual Revenue Per Gym"].str[1:]
df["Annual Revenue Per Member"]=df["Annual Revenue Per Member"].str[1:]
```

```python
df.drop(['Member Penetration Rate'],axis=1,inplace=True)
```

```python
df.rename(columns= {'Total # of Gyms':'TotalGyms','Total # of members':'TotalMembers','Total Revenue (USD)':'TotalRevenue(USD)'},inplace=True)
```

```python
df.astype('str')
df.replace(',','', regex=True, inplace=True)
df.head()
```

```python
df['TotalRevenue(USD)'] = df['TotalRevenue(USD)'].apply(pd.to_numeric)
df['Annual Revenue Per Gym'] = df['Annual Revenue Per Gym'].apply(pd.to_numeric)
df['Annual Revenue Per Member'] = df['Annual Revenue Per Member'].apply(pd.to_numeric)
```

```python
df.info()
```

## EDA

```python
df.head()
```

```python
ax=sns.barplot(data = df, x = 'Country',y= 'TotalMembers', order=df.sort_values('TotalMembers',ascending = False).Country)
sns.set(rc={'figure.figsize':(65,5)})
for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax=sns.barplot(data = df, x = 'Country',y= 'TotalRevenue(USD)', order=df.sort_values('TotalRevenue(USD)',ascending = False).Country)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax=sns.barplot(data = df, x = 'Country',y= 'Members Per Gym', order=df.sort_values('Members Per Gym',ascending = False).Country)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax=sns.barplot(data = df, x = 'Country',y= 'Annual Revenue Per Gym', order=df.sort_values('Annual Revenue Per Gym',ascending = False).Country)
sns.set(rc={'figure.figsize':(7,5)})
for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax=sns.barplot(data = df, x = 'Country',y= 'Annual Revenue Per Member',errwidth=0,order=df.sort_values('Annual Revenue Per Member',ascending = False).Country)
sns.set(rc={'figure.figsize':(95,5)})
for i in ax.containers:
    ax.bar_label(i,)
```
